## Idade

Para obter estatísticas descritivas de uma variável numérica em Python, é possível utilizar diretamente os métodos estatísticos disponibilizados pelos objetos Series do pandas. Essa abordagem é útil quando se deseja maior controle sobre quais medidas serão calculadas ou quando se pretende explicar cada estatística individualmente.

Supondo que os dados estejam armazenados em um DataFrame chamado df e que a variável numérica de interesse seja Idade, cada medida estatística pode ser obtida por meio de um método específico.

O valor mínimo e o valor máximo da variável podem ser calculados utilizando os métodos `min()` e `max()`, respectivamente. Essas medidas indicam os extremos observados no conjunto de dados.

A média aritmética é obtida com o método `mean()`, que calcula o valor médio da variável. Caso seja desejado limitar o número de casas decimais exibidas, a função `round()` pode ser utilizada para arredondamento.

A mediana, que representa o valor central da distribuição quando os dados estão ordenados, é calculada com o método `median()`. Essa medida é especialmente útil quando a distribuição apresenta valores extremos, pois é menos sensível a outliers do que a média.

O desvio padrão, que quantifica a dispersão dos dados em relação à média, é obtido por meio do método `std()`. Assim como na média, o valor pode ser arredondado para facilitar a leitura e a interpretação.

A moda, que corresponde ao valor mais frequente da variável, é calculada com o método `mode()`. Como esse método pode retornar mais de um valor em casos de múltiplas modas, é comum selecionar o primeiro elemento do resultado quando se deseja apenas uma moda representativa.

Além dessas medidas, é comum analisar percentis específicos da distribuição, que indicam o valor abaixo do qual uma determinada porcentagem dos dados se encontra. Percentis são particularmente úteis para entender a dispersão e a assimetria dos dados, bem como para identificar possíveis valores extremos.

Os percentis podem ser calculados utilizando o método `quantile()`. Por exemplo, o percentil 25 (primeiro quartil), o percentil 50 (que coincide com a mediana) e o percentil 75 (terceiro quartil) são frequentemente utilizados em análises exploratórias. Basta informar ao método o valor do percentil desejado em forma decimal, como 0.25, 0.50 ou 0.75. Também é possível calcular múltiplos percentis de uma única vez, passando uma lista de valores.

```
print(f"Mínimo: {df['Idade'].min()}")
print(f"Máximo: {df['Idade'].max()}")
print(f"Média: {round(df['Idade'].mean(),2)}")
print(f"Mediana: {df['Idade'].median()}")
print(f"Desvio padrão: {round(df['Idade'].std(),2)}")
print(f"Moda: {df['Idade'].mode().iloc[0]}")
print(f"3° quartil: {df['Idade'].quantile(0.75)}")
```

Dessa forma, utilizando métodos individuais como `min()`, `max()`, `mean()`, `median()`, `std()`, `mode()` e `quantile()`, é possível construir manualmente um conjunto completo de estatísticas descritivas para variáveis numéricas, compreendendo claramente o significado e a função de cada medida antes de recorrer a métodos mais automatizados, como o `describe()`, que reúne essas informações de forma resumida em uma única tabela.

## Sexo

Para variáveis categóricas, as estatísticas descritivas têm como objetivo principal resumir a distribuição das categorias e identificar padrões de frequência, em vez de medir tendência central ou dispersão numérica. Esse tipo de variável representa grupos, classes ou rótulos, como sexo, tipo de dor, presença ou ausência de uma condição, entre outros.

Supondo que os dados estejam armazenados em um DataFrame chamado df e que a variável categórica de interesse seja Sexo, as medidas descritivas mais comuns podem ser obtidas por meio de métodos específicos do pandas.

A principal medida de tendência central para variáveis categóricas é a moda, que corresponde à categoria mais frequente no conjunto de dados. Ela pode ser calculada utilizando o método `mode()`. Assim como no caso numérico, esse método pode retornar mais de uma categoria quando há empate de frequências. Quando se deseja apenas a categoria mais representativa, é comum selecionar o primeiro valor retornado.

Além da moda, a análise mais importante para variáveis categóricas é a contagem de valores, que indica quantas observações pertencem a cada categoria. Essa informação é obtida com o método `value_counts()`, que retorna a frequência absoluta de cada categoria presente na variável. Essa contagem permite avaliar o equilíbrio ou desbalanceamento das classes, aspecto fundamental em análises estatísticas e em aplicações de ciência de dados, como modelagem preditiva.

A partir da contagem de valores, também é possível derivar proporções ou porcentagens de cada categoria, o que facilita a comparação entre grupos, especialmente quando o tamanho da amostra é grande. Essas proporções ajudam a interpretar a representatividade de cada classe no conjunto de dados.

```
print(f"Moda: {df['Sexo'].mode().iloc[0]}")
print(f"Contagem de valores: {df['Sexo'].value_counts()}")
```

```
for col in df.select_dtypes(include="object"):
    print(f"\nVariável: {col}")
    print(df[col].value_counts())
```

## Visualização Gráfica

A escolha do tipo de gráfico deve ser guiada pelo tipo de variável que se deseja analisar e pelo objetivo da visualização. Um gráfico adequado facilita a interpretação dos dados, enquanto uma escolha inadequada pode induzir a conclusões equivocadas.

Para variáveis numéricas, o foco geralmente está na análise da distribuição, da tendência central, da dispersão e da relação entre variáveis. Quando o objetivo é compreender a distribuição de uma única variável numérica, gráficos como histograma e boxplot são os mais indicados. O histograma permite visualizar a forma da distribuição, identificar assimetrias e possíveis concentrações de valores, enquanto o boxplot resume a distribuição por meio de quartis, mediana e possíveis outliers.

Quando se deseja analisar a relação entre duas variáveis numéricas, o gráfico de dispersão é a escolha mais apropriada. Esse tipo de gráfico permite identificar padrões como correlação positiva, negativa ou ausência de relação, além de facilitar a detecção de agrupamentos e valores atípicos. Em alguns casos, uma linha de tendência pode ser adicionada para auxiliar a interpretação.

Para variáveis categóricas, o objetivo principal é comparar frequências ou proporções entre categorias. Nesses casos, os gráficos de barras são os mais utilizados. Cada barra representa uma categoria, e sua altura indica a frequência ou a proporção correspondente. Gráficos de barras são adequados tanto para variáveis categóricas nominais quanto ordinais, desde que a ordem das categorias seja respeitada quando ela existir.

Gráficos de setores, como o gráfico de pizza, também podem ser utilizados para variáveis categóricas, mas devem ser empregados com cautela. Eles são mais indicados quando há poucas categorias e quando o interesse está em visualizar proporções relativas. Em geral, gráficos de barras oferecem maior precisão na comparação entre categorias.

Quando se deseja analisar a relação entre duas variáveis categóricas, tabelas de contingência e gráficos de barras agrupadas ou empilhadas são as opções mais adequadas. Esses gráficos permitem comparar como as frequências de uma variável se distribuem dentro das categorias da outra, facilitando a identificação de associações visuais entre elas.

Em situações que envolvem uma variável numérica e uma variável categórica, gráficos que combinam ambos os tipos são recomendados. Boxplots ou violin plots, por exemplo, permitem comparar a distribuição de uma variável numérica entre diferentes categorias. Gráficos de barras com média ou mediana também podem ser utilizados, desde que acompanhados de medidas de variabilidade para evitar interpretações simplificadas.

Por fim, quando há uma dimensão temporal associada aos dados, gráficos de linhas são os mais indicados. Eles permitem visualizar a evolução de uma variável ao longo do tempo, identificar tendências, sazonalidades e mudanças de comportamento.

Em síntese, a seleção do tipo de gráfico deve considerar o tipo da variável (numérica ou categórica), o número de variáveis envolvidas e o objetivo da análise. Uma escolha consciente do gráfico é parte fundamental da análise exploratória de dados e contribui diretamente para a clareza e a qualidade das conclusões extraídas.

# 3. Associação de Variáveis Categóricas

O objetivo é avaliar se existe relação entre as variáveis qualitativas ou categóricas estudadas, além do grau de associação entre elas. Isto pode ser feito por meio de tabelas de distribuições de frequências, medidas-resumo como o qui-quadrado (utilizado para variáveis nominais e ordinais), o coeficiente Phi, o coeficiente de contingência e o coeficiente V de Cramer (para variáveis nominais), e o coeficiente de Spearman (para variáveis ordinais), além de representações gráficas como mapas perceptuais provenientes da análise de correspondência


Fonte: Fávero, Luiz Paulo, and Patrícia Belfiore. Manual de análise de dados: estatística e modelagem multivariada com Excel®, SPSS® e Stata®. Elsevier Brasil, 2017.

### Tabelas de Contingência

Tabela de contingência exibe, de forma conjunta, as frequências absolutas ou relativas das categorias da variável X, representada no eixo das abscissas, e da variável Y, representada no eixo das ordenadas.

Fonte: Fávero, Luiz Paulo, and Patrícia Belfiore. Manual de análise de dados: estatística e modelagem multivariada com Excel®, SPSS® e Stata®. Elsevier Brasil, 2017.

```
df_avc = df.select_dtypes(include="object")
df_avc
```

```
from scipy.stats import chi2_contingency
```

```
obs = pd.crosstab(df_avc["Sexo"], df_avc["DorPeito"])
chi2, p_value, gl, expected = chi2_contingency(obs)
```

```
chi2, p_value, gl, expected
```

```
alpha = 0.05
if p_value < alpha:
    print("Existe associação estatisticamente significativa")
else:
    print("Não há evidência de associação estatisticamente significativa")
```
