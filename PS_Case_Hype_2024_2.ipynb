Modelos escolhidos: Random Forest e KNN

Ambos são bons modelos para classificação, mas que possuem diferenças na complexidade e nos hiperparâmetros.

O KNN, por exemplo, é bem mais simples de entender pois é baseado no cálculo da distância entre as variáveis. Entranto, ele pode demorar mais para calcular os pontos em um conjunto muito grande. Ainda assim, ele apresenta uma acurácia relativamente boa para sua simplicidade.

O Random Forest, por outro lado, é um modelo mais complexo e com vários hiperparâmetros que podem ser modificados para melhorar os resultados. Sua lógica se baseia na construção de diversas árvores de decisão da combinação delas para chegar no melhor resultado. Como compensação, ele é mais eficiente e pode obter acurácias melhores quando se trata de grande volume de dados.

A princípio, testaremos os dois modelos para decidir qual o melhor nesse caso.
